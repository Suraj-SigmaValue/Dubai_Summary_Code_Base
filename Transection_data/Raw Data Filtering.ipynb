```python
df['Year']=df['instance_date'].str[6:]
df['Year']
```

```python
# First ensure instance_date is in datetime format
df['instance_date'] = pd.to_datetime(df['instance_date'], format='%d-%m-%Y', errors='coerce')

# Create Year column
df['Year'] = df['instance_date'].dt.year

# Create Quarter column in format "Q1 2022"
df['Quarter'] = 'Q' + df['instance_date'].dt.quarter.astype(str) + ' ' + df['instance_date'].dt.year.astype(str)

# Alternative: Create separate Quarter and Year Quarter columns if needed
df['Quarter_Year'] = df['instance_date'].dt.to_period('Q')
```

```python
df
```

```python
flat_types = {
    "Flat", "Villa", "Stacked Townhouses", "Unit"
}

shop_types = {
    "Shop", "Store", "Show Rooms"
}

office_types = {
    "Office", "Clinic", "Workshop", "Warehouse"
}

others = {
    "Hotel Apartment",
    "Hotel Rooms",
    "Gymnasium",
    "Sized Partition",
    "Hotel",
    "Building"
}
def categorize_property(sub_type):
    if sub_type in flat_types:
        return "Flat"
    elif sub_type in shop_types:
        return "Shop"
    elif sub_type in office_types:
        return "Office"
    else:
        return "Others"

    
df['property_type'] = df['property_sub_type_en'].apply(categorize_property)
```

```python
df=df[df['property_type_en']!='Land']
# df=df[df['Year'].isin([2020, 2021, 2022, 2023, 2024, 2025, 2026])]
df
```

```python
df['procedure_area_sqft']=df['procedure_area']*10.7639
df['sqft_sale_price']=df['meter_sale_price']/10.7639
```

```python
import re
import pandas as pd

def map_to_bhk(val):
    if pd.isna(val):
        return None

    val = val.strip().upper()

    # Group large BHKs first
    if val in { "6 B/R","7 B/R", "8 B/R", "9 B/R"}:
        return ">5 B/R"

    # Match normal BHKs (1–6)
    match = re.match(r"(\d+)\s*B/R", val)
    if match:
        bhk = int(match.group(1))
        if bhk <= 5:
            return f"{bhk} B/R"

    # Special residential cases
    if val == "SINGLE ROOM":
        return "<1 B/R"

    else:
        return "Commercial"

df["BHK"] = df["rooms_en"].apply(map_to_bhk)
```
